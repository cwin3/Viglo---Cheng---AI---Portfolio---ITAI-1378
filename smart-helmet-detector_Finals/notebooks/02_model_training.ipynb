```python
# 02_model_training.ipynb

# 1. Setup
!pip install ultralytics
from ultralytics import YOLO
import os

# ---------------------------------------------------------
# 2. LOAD DATASET
# ---------------------------------------------------------
# If you are in a new Colab session, you might need to download the data again
# (Paste the Roboflow snippet here again if needed)
# ...
!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="KCpMcoMPTL2pLqR6hpSG")
project = rf.workspace("group-itai-1378-project").project("helmets-ewjdr-p2lu5")
version = project.version(1)
dataset = version.download("yolov8")

# Locate the data.yaml file
data_yaml_path = f"{dataset.location}/data.yaml"
print(f"Data config found at: {data_yaml_path}")

# ---------------------------------------------------------
# 3. TRAIN MODEL
# ---------------------------------------------------------
# We use 'yolov8n.pt' (Nano) because it's fast for real-time demos
model = YOLO('yolov8n.pt')

print("Starting training...")
# Train for 25-50 epochs.
# imgsz=640 is standard.
model.train(
    data=data_yaml_path,
    epochs=50,
    imgsz=640,
    plots=True,
    batch=16,
    name='helmet_detector_model' # This saves results to runs/detect/helmet_detector_model
)

print("Training Complete!")

# ---------------------------------------------------------
# 4. EXPORT
# ---------------------------------------------------------
# The best model is saved automatically, but let's copy it to a clear location
# output_dir = runs/detect/helmet_detector_model/weights/best.pt

import shutil

destination = "/content/best.pt" # Easier to download from here
source = "runs/detect/helmet_detector_model/weights/best.pt"

shutil.copy(source, destination)
print(f"Model saved to {destination} - DOWNLOAD THIS FILE NOW!")
```
